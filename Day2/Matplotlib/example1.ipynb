```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("D:\Machine Learning\Day2\Matplotlib\diamonds.csv")
```

```python
df.head()
```

Basic Plots using Matplotlib

```python
plt.plot(df['carat'],df['price'])
plt.title('carat vs Price Line Plot')
plt.xlabel('Carat')
plt.ylabel("Price")
```

```python
plt.scatter(df['carat'],df['price'])
plt.title('carat vs Price Scatter Plot')
plt.xlabel('Carat')
plt.ylabel("Price")
```

```python
# Bar Plot 
plt.bar(df['cut'].unique(),df.groupby('cut')['price'].mean())
plt.title('Average price by Cut')
plt.xlabel('Cut')
plt.ylabel('Avarage Price')
plt.show()
```

Intermediate Plots using matplotlib

```python
plt.hist(df['price'],bins=20,edgecolor='black')
plt.title('Distribution of Prices')
plt.xlabel("Price")
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot 
plt.boxplot(df['carat'],vert=False) # Horizontal 
# plt.boxplot(df['carat'],vert=True) # Vertical 
plt.title("Box Plot of Carat")
plt.xlabel('Carat')
plt.show()
```

```python
# Pie Chart
cut_counts = df['cut'].value_counts()
plt.pie(cut_counts, labels = cut_counts.index, autopct = '%1.1f%%',startangle=90)
plt.title("Distribution of Cut Quantity")
plt.show()
```

Advanced Ploting using Matplotlib

```python
correlation = df[['carat','depth','table','price','x','y','z']].corr()
correlation
```

```python
# Heatmap of correlation Matrix
correlation = df[['carat','depth','table','price','x','y','z']].corr()
plt.imshow(correlation,cmap='coolwarm',interpolation='none')
plt.colorbar()
plt.xticks(range(len(correlation)),correlation.columns,rotation = 80)
plt.yticks(range(len(correlation)),correlation.columns)
plt.title("Heatmap of Correlation Matrix")
plt.show()
```

```python
sns.heatmap(correlation,cmap='coolwarm',annot=True)
```

```python
# subplots
fig , axes = plt.subplots(1,2,figsize =(12,5))
axes[0].hist(df['carat'],bins = 10, edgecolor = 'black')
axes[0].set_title('Histogram of Carat')
axes[1].hist(df['price'],bins =10, edgecolor ='black')
axes[1].set_title('Histogram of Prices')
plt.show()
```

```python
fig , axes = plt.subplots(2,2,figsize =(12,10))
axes[0,0].hist(df['carat'],bins = 10, edgecolor = 'black')
axes[0,0].set_title('Histogram of Carat')
axes[0,1].hist(df['price'],bins =10, edgecolor ='black')
axes[0,1].set_title('Histogram of Prices')

axes[1,0].scatter(df['carat'],df['price'])
axes[1,0].set_title("Scatter Plot")
axes[1,1].bar(df['cut'].unique(), df.groupby('cut')['price'].mean())
axes[1,1].set_title('Average Price by cut')

plt.show()
fig.tight_layout()
```

```python
# Enhancing Plots with Seaborn for Report-Ready Visualization
```

```python
# Setting Seaborn Styles
sns.set_style("darkgrid") # Options : White , Dark , whitegrid, darkgrid, ticks
sns.set_context('notebook') # Options : paper, Notebook, Talk, poster

sns.scatterplot(x='carat',y='price',data=df)
plt.title("Display the Graph")
plt.show()
```

```python
sns.set_context('paper', font_scale=1)
sns.boxplot(x='cut',y='price',data = df)
plt.title("Box plot with Talk context and Larger Font Scale")
plt.show()
```

```python
# Using a predifined color pallette
sns.set_context('notebook')
sns.set_palette('deep')  # Options : deep , muted, pastel, dark, colorblind
sns.violinplot(x='color',y='price',data=df , hue='cut')
plt.title("Using color pallters and violin graph")
plt.show()
```

```python
# Creating and Using custom color palette
custom_palette = sns.color_palette(['#000000','#0000FF','#00FFFF',"#FF00FF","#808080","#00FF00","#7DF9FF"])
sns.set_palette(custom_palette)
sns.countplot(x='cut', data=df, hue='color')
plt.title("Count Plot with custom color palette")
plt.show()
```

```python

# Adding grid lines to the background
sns.set_style('whitegrid')
sns.scatterplot(x='carat',y='price',data=df)
plt.grid(True, linestyle = '--',alpha = 0.5)
plt.title("Scatter Plot with Gridlines")
plt.show()
```

```python
sns.set_style('whitegrid')
ax = sns.histplot(df['price'], bins=10, edgecolor = 'black',color='green')
sns.despine() # Removes top and Right spine
plt.title("Histogram with spines removed")
plt.show()
```

```python
#Adding  Annotation to a scatter plot
sns.scatterplot(x='carat',y='price',data=df)
plt.title("Scatter plot with Annotation")
for i in range(len(df)):
    plt.annotate(f"${df['price'][i]}", (df['carat'][i],df['price'][i]),fontsize = 8, alpha = 0.7) 
plt.show()
```

```python
# Cutomizing the legend
figure = plt.figure(figsize=(12, 3))
sns.boxplot(x='cut', y = 'price', data = df)
plt.legend(title = "Cut Types", labels = df['cut'].unique(), loc = 'upper right', frameon = True)
plt.title("Box plot with cutsomizing legend")
plt.show()
```

```python
# Cutomizing the legend
figure = plt.figure(figsize=(12, 3))
sns.boxplot(x='cut', y = 'price', data = df)
plt.legend(title = "Cut Types", labels = df['cut'].unique(), loc = 'upper left', bbox_to_anchor = (1,1), frameon = True)
plt.title("Box plot with cutsomizing legend")
plt.show()
```

```python
# Using FacetGrid for multiple histograms based on cut
g = sns.FacetGrid(df, col='cut', col_wrap= 3, height=4 )
g.map(plt.hist, 'price', bins =10, edgecolor = 'black')
g.set_titles("{col_name} Cut")
plt.show()
```

```python
fig = plt.figure(figsize=(12,6))

g = sns.FacetGrid(df, hue='color', height = 5)
g.map(plt.scatter, 'carat', 'price').add_legend()
plt.title('FacetGrid Scatter plot with Hue')
plt.show()
```

```python
# Rotating x - axis labels for better readability
sns.barplot(x='color', y='price', data = df)
plt.xticks(rotation = 45)
plt.title("Bar Plot with rotated x- axis lables")
plt.show()
```

```python
# Adjusting the y-axis scale
sns.histplot(df['price'],bins=10,edgecolor = 'black')
plt.yscale('log')
plt.title('Histogram with Logarithmic Y-axis Scale')
plt.show()
```

```python
sns.histplot(df['price'],bins = 10, edgecolor='black')
# plt.yscale('log')
plt.title("histogram with log y - axis Scale")
plt.show()
```

```python
# applyinng the 'ticks' theme

g = plt.figure(figsize=(12,3))
sns.set_theme(style='ticks',palette = 'muted',context='notebook')
sns.lineplot(x='carat',y='price',data=df)
sns.despine()
plt.title('Line plot with ticks theme and muted Palette')
plt.show()
```

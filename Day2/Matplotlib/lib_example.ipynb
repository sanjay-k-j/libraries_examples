```python
# Customizing plots in Seaborn and Matplolib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('diamonds.csv')
print("Number of rows before sampling : ",df.shape[0])
print("Number of columns before sampling : ",df.shape[1])

print(f"{50*'-'}")

df = df.sample(5000)
print(f"Number of rows after sample: ",df.shape[0])
print(f"Number of Columns after sampling : ",df.shape[1])
df.head()
```

```python
# Setting Title
sns.scatterplot(x='carat',y='price',data=df)
plt.title("Scatter plot of Carat VS Price")
plt.show()
```

```python
# Setting Axes labels
sns.scatterplot(x='carat',y='price',data=df)
plt.xlabel('Carat of diamond')
plt.ylabel('Price (in USD)')
plt.show()
```

```python
# Customizing the legend
sns.scatterplot(y='price',x='carat',hue='cut',data=df)
plt.legend(title = 'Diamond Plot', loc = 'upper right')
plt.show()
```

```python
#Customizing Ticks 

sns.histplot(df['price'],bins=10, edgecolor = 'blue')
plt.xticks(rotation = 45,fontsize = 15)
plt.yticks(fontsize = 10)
plt.show()
```

```python
# creating subplots

fig, axes = plt.subplots(1,2, figsize = (12,5))
sns.histplot(df['carat'],bins=10, edgecolor = 'black', ax = axes[0],kde = True)
axes[0].set_title("Histogram of Carat")
sns.histplot(df['price'],bins=10,edgecolor = "black",ax = axes[1], kde=True)
axes[1].set_title("Histogram of Price")
plt.show()
```

```python
# Adding annotations 
df.reset_index(drop=True , inplace=True)
sns.scatterplot(x='carat', y='price',data = df)
plt.title("Scatter Plot with Annotations")
for i in range(len(df)):
    plt.annotate(f'${df['price'][i]}',(df['carat'][i],df['price'][i]), fontsize = 8)
plt.show()
```

```python
# Cusomizing Color Palettes

sns.set_palette('deep')
sns.boxplot(x='cut', y='price',data=df,hue='clarity')
plt.title('Box Plot with muted color Pallte')
plt.show()
```

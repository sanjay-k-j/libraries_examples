```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go 
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('D:\Machine Learning\Day2\Matplotlib\diamonds.csv')
df = df.sample(frac = 0.1, random_state=10)
```

```python
# Video plot
fig = px.scatter(df,
                 x= 'carat', y = 'price', animation_frame = 'cut',color='cut',
                 size = 'price', facet_col='color',facet_row='clarity',
                 title='Diamond Price vs Carat weight'
                 )

# size of figure 
fig.update_layout(height = 800, width = 600)
fig.show()

fig.write_html('D:\Machine Learning\Day2\Saved HTML\diamond_price_animated.html')
```
